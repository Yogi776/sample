```python
from faker import Faker
fake = Faker('de_DE')
```

```python
 fake.address()
```

```python
def email():
    return  fake.ascii_email()
```

```python
def address():
    return fake.address()
```

```python
def mobile():
    return fake.phone_number()
```

```python
df1['email'] = df1.apply(lambda x: email(), axis = 1)
df1['address'] = df1.apply(lambda x: address(), axis = 1)
df1['mobile'] = df1.apply(lambda x: mobile(), axis = 1)
```
